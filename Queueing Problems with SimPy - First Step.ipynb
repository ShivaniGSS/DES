```python
# Class to hold global variable for easy access
class G:
    DURATIONS = 8 * 60   # Hours of simulation run in minutes
    ITERATIONS = 100     # Number of times to run simulation
    ARRIVALS = 2         # Mean rate of new customer arrival
    NUM_TELLERS = 1      # Number of servers
    SERVETIME = 1.5
    customer_data = {}


def customer(env, name, agent, time_to_arrival, time_with_agent):
     # Simulate customer arriving
    yield env.timeout(time_to_arrival)

     # Request an agent
    #print('%s requests an agent at %d' % (name, env.now))
    gets_there = env.now
    with agent.request() as req:
        yield req

         # Work with the agent
        #print('%s sees agent at %s' % (name, env.now))
        works_with = env.now
        yield env.timeout(time_with_agent)
        #print('%s leaves the agent at %s' % (name, env.now))
        leaves = env.now
        G.customer_data[name] = [gets_there, works_with, leaves, time_with_agent]
        
#Create the Customer arrival times, time and the processes before simulation. Create more
#than needed and will just not use all. In the next case I'll create a generator 
#function to create during the simulation.

def make_processes():
    arrival_times = []
    times_with_agent = [np.random.exponential(G.SERVETIME) for i in range(100)]
    time = 0
    for i in range(300):
        time += np.random.exponential(2)
        arrival_times.append(time)
    for i in range(300):
        env.process(customer(env, 'Customer %d' % i, agent, arrival_times[i], np.random.exponential(G.ARRIVALS)))
        
#Our resource in this case is the server/teller
sim_results = {}
queue_times = {}
system_times = {}
agent_busy = []
agent_idle = []
for i in range(G.ITERATIONS):
    G.customer_data = {}
    env = simpy.Environment()
    agent = simpy.Resource(env, capacity=G.NUM_TELLERS)
    make_processes()
    env.run(until=env.timeout(480))
    time_sim_ends = env.now
    customer_df = pd.DataFrame(G.customer_data).T.rename(columns = {0:"arrives", 1: "service_starts", 2: "leaves", 3: "time_with_agent"})
    sim_results[i] = customer_df
    
    time_agent_busy = np.sum(customer_df["time_with_agent"])
    agent_busy.append(time_agent_busy)
    
    time_agent_idle = time_sim_ends - time_agent_busy
    agent_idle.append(time_agent_idle)
    
    time_in_queue = pd.DataFrame(customer_df["service_starts"] - customer_df["arrives"])
    queue_times[i] = time_in_queue
    
    time_in_system = pd.DataFrame(customer_df["leaves"] - customer_df["arrives"])
    system_times[i] = time_in_system
        
```

```python
from simpy.resources.resource import Resource, Request, Release
from simpy.core import BoundClass
from simpy.resources.base import BaseResource

""""" First need to modify the resource code to allow change of number of agents"""
class VariableResource(BaseResource):
    def __init__(self, env, capacity):
        super(VariableResource, self).__init__(env, capacity)
        self.users = []
        self.queue = self.put_queue

    @property
    def count(self):
        return len(self.users)

    request = BoundClass(Request)
    release = BoundClass(Release)

    def _do_put(self, event):
        if len(self.users) < self.capacity:
            self.users.append(event)
            event.usage_since = self._env.now
            event.succeed()

    def _do_get(self, event):
        try:
            self.users.remove(event.request)
        except ValueError:
            pass
        event.succeed()
    
    # This is the addition. If each resource had it's own queue then the handling of this change
    # would be a bit messier depending on we decide the queue closes (redistribute costumers or finish in queue)
    def _change_capacity(self, capacity):
        self._capacity = capacity
        #event.succeed()
```

### Simple case adding some structure

```python
# Define Global variables in a class
class G:
    DURATION = 8 * 60   # Hours of simulation run in minutes
    ITERATIONS = 100     # Number of times to run simulation
    #ARRIVALS = [2, 1.6, 1.3, 1.05, 1, 1.1, 1.4, 1.8]         # Mean time in minutes between arrivals - exponential distribution
    #SERVERS = [1, 1, 2, 2, 3, 3, 2, 2]          # Number of servers working
    #rate_of_change = 60
    ARRIVALS = 2
    NUM_TELLERS = 1
    SERVETIME = 1.5
    customer_data = {}
    arrival_times = []
    
    
class Bank(object):
    """"" A bank has a limited number of tellers to serve
    customers.
    
    Customers must request a server and when they get one wil be 
    served on a varying exponential time scale"""
    
    def __init__(self, env, num_tellers, servicetime):
        self.env = env
        self.teller = VariableResource(env, num_tellers)
        self.servicetime = servicetime
    
    def serve(self, name):
        """"" The process of the customer interacting with the teller"""     
        #print('%s sees agent at %.2fs' % (name, env.now))
        yield self.env.timeout(np.random.exponential(self.servicetime))
        #print('%s leaves the agent at %.2fs' % (name, env.now))
        
        
def customer(env, name, bank, data):
        # Simulate customer arriving
        #print('%s arrives at the Bank at %.2f.' % (name, env.now))
        arrives = env.now
        with bank.teller.request() as request:
            yield request
            
            #print('%s sees the teller at %.2f.' % (name, env.now))
            service_begins = env.now
            
            yield env.process(bank.serve(name))
            #print('%s leaves the teller at %.2f.' % (name, env.now))
            leaves = env.now
            
            #store data
            data[name] = [arrives, service_begins, leaves, leaves - service_begins]
            
def setup(env, num_tellers, servicetime, arrival_rate, data):
    """Create a bank creates customers approx. every ``t_inter`` minutes."""
    # Create the bank
    bank = Bank(env, num_tellers, servicetime)
    
    i = 0
    # Create customers while the simulation is running
    while True:
        yield env.timeout(np.random.exponential(arrival_rate))
        i += 1
        env.process(customer(env, 'Customer %d' % i, bank, data))
    
# Setup and start the simulation
#print('Bank')
# Create an environment and start the setup process
for i in range(G.ITERATIONS):
    data = {}
    env = simpy.Environment()
    env.process(setup(env, G.NUM_TELLERS, G.SERVETIME, G.ARRIVALS, data))
    env.run(until=G.DURATION)
    time_sim_ends = env.now
    
    customer_df = pd.DataFrame(data).T.rename(columns = {0:"arrives", 1: "service_starts", 2: "leaves", 3: "time_with_agent"})
    sim_results[i] = customer_df
    
    time_agent_busy = np.sum(customer_df["time_with_agent"])
    agent_busy.append(time_agent_busy)
    
    time_agent_idle = time_sim_ends - time_agent_busy
    agent_idle.append(time_agent_idle)
    
    time_in_queue = pd.DataFrame(customer_df["service_starts"] - customer_df["arrives"])
    queue_times[i] = time_in_queue
    
    time_in_system = pd.DataFrame(customer_df["leaves"] - customer_df["arrives"])
    system_times[i] = time_in_system
```

## Time Dependency on Resources and Arrival time

```python
# Define Global variables in a class
class G:
    DURATIONS = 8 * 60   # Hours of simulation run in minutes
    ITERATIONS = 100     # Number of times to run simulation
    ARRIVALS = [2, 1.6, 1.3, 1.05, 1, 1.1, 1.4, 1.8]         # Mean time in minutes between arrivals - exponential distribution
    TELLERS = [1, 1, 2, 2, 3, 3, 2, 2]          # Number of servers working
    SERVETIME = 1.5
    customer_data = {}
    

class Bank(object):
    """"" A bank has a limited number of tellers to serve
    customers.
    
    Customers must request a server and when they get one wil be 
    served on a varying exponential time scale"""
    
    def __init__(self, env, num_tellers, servicetime):
        self.env = env
        self.teller = VariableResource(env, num_tellers)
        self.servicetime = servicetime
    
    def serve(self, name):
        """"" The process of the customer interacting with the teller"""     
        #print('%s sees agent at %.2fs' % (name, env.now))
        yield self.env.timeout(np.random.exponential(self.servicetime))
        #print('%s leaves the agent at %.2fs' % (name, env.now))
        
        
def customer(env, name, bank, data):
        # Simulate customer arriving
        #print('%s arrives at the Bank at %.2f.' % (name, env.now))
        arrives = env.now
        with bank.teller.request() as request:
            yield request
            
            #print('%s sees the teller at %.2f.' % (name, env.now))
            service_begins = env.now
            
            yield env.process(bank.serve(name))
            #print('%s leaves the teller at %.2f.' % (name, env.now))
            leaves = env.now
            num_tellers = bank.teller._capacity
            #store data
            data[name] = [arrives, service_begins, leaves, num_tellers]
            
#Activity generator function            
def change_tellers(env, bank, num):
    shift_change_time = env.now
    #print(f"Shift change to {num} tellers")
    with bank.teller.request() as req:
        #Freeze until teller is free
        if num < bank.teller._capacity:
            yield req
        # change number of tellers
        bank.teller._change_capacity(num)
        #print(bank.teller._capacity, env.now)

# Shift change generator function
def shift_change(env, bank):
    t=0
    i=0
    for i in range(len(G.TELLERS)):
        yield env.timeout(t)
        t=60
        env.process(change_tellers(env, bank, G.TELLERS[i]))
        #print(G.TELLERS[i])
        
def get_arrival_rate(env):
    i = int(np.floor(env.now/60))
    return G.ARRIVALS[i]


def setup(env, bank, data):
    """Create a bank creates customers approx. every ``t_inter`` minutes."""
    i = 0
    # Create customers while the simulation is running
    while True:
        yield env.timeout(np.random.exponential(get_arrival_rate(env)))
        i += 1
        env.process(customer(env, 'Customer %d' % i, bank, data))
            

        
# Setup and start the simulation
#print('Bank')
# Create an environment and start the setup process
data = {}
sim_results_comp = {}
for i in range(G.ITERATIONS):
    env = simpy.Environment()
    bank = Bank(env, G.TELLERS[0], G.SERVETIME)
    env.process(shift_change(env, bank))
    env.process(setup(env, bank, data))
    env.run(until=G.DURATIONS)

    customer_df = pd.DataFrame(data).T.rename(columns = {0:"arrives", 1: "service_starts", 2: "leaves", 3: "num_tellers"})
    
    sim_results_comp[i] = customer_df
    
    time_in_queue = pd.DataFrame(customer_df["service_starts"] - customer_df["arrives"])
    queue_times[i] = time_in_queue
    
    time_in_system = pd.DataFrame(customer_df["leaves"] - customer_df["arrives"])
    system_times[i] = time_in_system
```
